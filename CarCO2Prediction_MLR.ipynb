```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/DE/ML/Datasets/AIML_DST/data.csv")
data.head()
```

```python
 x = data.iloc[:, 2:-1].values
 y = data[['CO2']]
 from sklearn.model_selection import train_test_split
 xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 32)
```

```python
sns.scatterplot(x = 'Volume', y = 'CO2', data = data, color = 'skyblue')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
print(model.intercept_, model.coef_)
```

```python
ypred = model.predict(xtest)

from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
print("MAE: ", mean_absolute_error(ypred, ytest))
print("MSE: ", mean_squared_error(ypred, ytest))
print("RMSE: ", root_mean_squared_error(ypred, ytest))
print("R2 Score: ", r2_score(ypred, ytest))
```
